# Filling
data.fillna()

```python
missing_values = ['NAN','V','P','NA']  
#standardized missing values as pandas format
iris = pd.read_csv('iris.csv',na_values = missing_values)
```

```python
iris
```

# Filling with mean/ median / mode / min / max / sum 

```python
#filling with median
iris['sepal.length'].fillna(iris['sepal.length'].mode() , inplace= True)
```

```python
iris
```

```python
# filling with constant value
iris['sepal.width'].fillna( 1.2 , inplace= True) 
```

```python
iris
```

```python
# filling with forward value 
iris['petal.length'].fillna( method= 'ffill', inplace= True)
```

```python
iris
```

```python
# filling with backward value 
iris['variety'].fillna( method= 'bfill', inplace= True)
```

```python
iris
```

```python
#  filling with linear interpolation 
#  method = 'linear'
#  method = 'polynomial'
#  method = many method check pandas site

iris['petal.width'].interpolate(method= 'linear', inplace= True)
```
